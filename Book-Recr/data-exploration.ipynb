```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
import pandas as pd
```

```python
books = pd.read_csv(f"{path}/Books.csv")
```

```python
books
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ax = plt.axes()
sns.heatmap(books.isna().transpose(),cbar=False,ax=ax)

plt.xlabel("Columns")
plt.ylabel("Missing values")
plt.show()
```

```python
import numpy as np
books["missing_description"] = np.where(books["description"].isna(),1,0)
books["age"] = 2025 - books["published_year"]
```

```python
columns_of_interest=["num_pages" , "age","missing_description","average_rating"]

correlation_matrix = books[columns_of_interest].corr(method="spearman")
sns.set_theme(style="white")
plt.figure(figsize=(8, 6))
heatmap = sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm',cbar_kws={"label": "spearman corr"})
heatmap.set_title("Correlation Matrix HeatMap")
plt.show()
```

```python
book_missing = books[~(books["description"].isna()) & ~(books["num_pages"].isnull()) 
       & ~(books["average_rating"].isnull())
       & ~(books["published_year"].isnull())
       ]
book_missing
```

```python
book_missing["words_in_des"]= book_missing['description'].str.split().str.len()
book_missing
```

```python
book_missing.loc[book_missing["words_in_des"].between(1,4) , "description"]
```

```python
book_missing.loc[book_missing["words_in_des"].between(25,34) , "description"]
```

```python
book_missing_25_words = book_missing[book_missing["words_in_des"]>=25]
book_missing_25_words
```

```python
book_missing_25_words["title_and_subtitle"] = (
    np.where(book_missing_25_words["subtitle"].isna(), book_missing_25_words["title"],
             (book_missing_25_words[["title" , "subtitle"]].astype(str).agg(":".join, axis=1)
              )
))
book_missing_25_words
```

```python
book_missing_25_words["tag_description"] = book_missing_25_words[["isbn13", "description"]].astype(str).agg(" ".join, axis=1)
```

```python
book_missing_25_words
```

```python
(
    book_missing_25_words.drop(["subtitle" , "missing_description" , "age" , "words_in_des"], axis=1)
    .to_csv("books_cleaned.csv",index=False)
)
```
