```python
import pandas as pd
books = pd.read_csv('books_cleaned.csv')
books
```

```python
books["tag_description"]
```

```python
books["tag_description"].to_csv("tagged_des.txt" , sep="\n" ,index=False , header = False)
```

```python
raw_documents = TextLoader("tagged_des.txt" , encoding="utf8").load()
text_splitter = CharacterTextSplitter(chunk_size=0, chunk_overlap=0,separator= "\n")
documents = text_splitter.split_documents(raw_documents)
```

```python
documents[0]
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings
huggingface_embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
db_books = Chroma.from_documents(documents, embedding = huggingface_embeddings)
```

```python
query = "a book to teach children about nature"
docs = db_books.similarity_search(query , k = 10)
docs
```

```python
import re

def retrieve_semantic_rec(query: str, top_k: int = 10) -> pd.DataFrame:
    recs = db_books.similarity_search(query, k=50)
    
    books_list = []
    for rec in recs:
        # Extract ISBN digits from the beginning of the content
        match = re.match(r'^"?(\d+)', rec.page_content.strip())
        if match:
            books_list.append(int(match.group(1)))
    
    return books[books["isbn13"].isin(books_list)].head(top_k)
```

```python
retrieve_semantic_rec("a book to teach children about nature")
```
